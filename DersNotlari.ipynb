```python
indirme_listesi = ['XU100.IS', 'TRY=X']
Veri_2023 = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start='2023-01-01', end='2023-04-01')
  data = data["Adj Close"].rename(i)
  Veri_2023 = pd.concat([Veri_2023, data], axis=1)

Veri_2023
```

```python
Veri_2023.columns = ["Kapanış (TRY)","USD/TRY"]
Veri_2023
```

```python
Veri_2023 = Veri_2023[Veri_2023["Kapanış (TRY)"] > 0]
```

```python
Veri_2023
```

```python
Veri_2022 = B100K[["Kapanış (TRY)","USD/TRY"]]
Veri_2022
```

```python
Veri = pd.concat([Veri_2022, Veri_2023], axis=0)
Veri
```

```python
Veri["Normalize"] = Veri["Kapanış (TRY)"]/Veri.iloc[0,0]
Veri["Normalize USD/TRY"] = Veri["USD/TRY"]/Veri.iloc[0,1]
Veri["Degisim"] = Veri["Normalize"].diff(1)
Veri["Degisim-Yonu"] = Veri["Degisim"].map((lambda x : "-" if x<0 else "+"))
Veri["Degisim"] = (Veri["Degisim"]/Veri["Normalize"].shift(1))*100
Veri
```

iloc yerine loc ile gösterimi

```python
Veri["Normalize"] = Veri["Kapanış (TRY)"]/Veri.iloc[0,0]
Veri["Normalize USD/TRY"] = Veri["USD/TRY"]/Veri.loc[Veri.index[0], "USD/TRY"]
Veri["Degisim"] = Veri["Normalize"].diff(1)
Veri["Degisim-Yonu"] = Veri["Degisim"].map((lambda x : "-" if x<0 else "+"))
Veri["Degisim"] = (Veri["Degisim"]/Veri["Normalize"].shift(1))*100
Veri
```

grafik çizdirmek

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=Veri.index, y=Veri["Normalize"], name="Bist100 Endeksi Normalize Değeri"))
fig.add_trace(go.Scatter(x=Veri.index, y=Veri["Normalize USD/TRY"], name="Normalize USD/TRY"))

fig.update_layout(title="Bist100 Endeksi Normalize Değeri",
                  xaxis_title="Tarih", xaxis_dtick="M1",
                  yaxis_title= "Kapanış TRY")
```

```python
url = "https://www.tcmb.gov.tr/wps/wcm/connect/TR/TCMB+TR/Main+Menu/Istatistikler/Enflasyon+Verileri/Tuketici+Fiyatlari"
```

```python
response = requests.get(url)
response.text
```

```python
soup = bs(response.text, "html.parser")
```

```python
soup
```

```python
soup = bs(response.text, 'html.parser')
table = soup.find('table', {'summary' : 'Fiyat Endeksi (Tüketici Fiyatları)(2003=100)'})
table
```

```python
print(table)
```

```python
tufe = pd.read_html(str(table))
tufe = pd.DataFrame(tufe[0])
```

```python
tufe
```

```python
tufe = tufe[["Unnamed: 0", "TÜFE (Aylık % Değişim)"]]
tufe.columns = ["index", "tufe%"]
tufe
```

```python
tufe["index2"] = tufe["index"].str[-4:]
tufe
```

```python
tufe["index2"] = tufe["index"].str[:2]
tufe
```

```python
tufe["index"] = tufe["index"].str[-4:] + "-" + tufe["index"].str[:2]
tufe
```

```python
tufe.set_index("index", inplace=True)
```

```python
tufe
```

```python
tufe.reset_index(inplace=True)
```

```python
tufe
```

```python
tufe["tufe%"] = tufe["tufe%"]/100
tufe
```

```python
tufe = tufe[tufe.index.str[:4] > "2021"]
tufe
```

```python
Veri_ay_sonu
```

```python
Veri_enflasyon = pd.concat([Veri_ay_sonu, tufe], axis=1)
Veri_enflasyon
```

```python
Veri_enflasyon["Degisim Yuzdesi Enflasyon"] = ((Veri_enflasyon["Degisim"]+1)/(Veri_enflasyon["tufe%"]+1))-1
Veri_enflasyon["Degisim-Yon-Enflasyon"] = Veri_enflasyon["Degisim Yuzdesi Enflasyon"].map((lambda x : "-" if x<0 else "+"))
Veri_enflasyon
```

Reel Getiri = ((1+değişim%)/(1+tufe%))-1

```python
px.bar(Veri_enflasyon, y = "Degisim Yuzdesi Enflasyon", color = "Degisim-Yon-Enflasyon", color_discrete_map={"+": "green", "-": "red"},
       text_auto=".0%")
```
